### Find outliers through boxplot 

```python
columns_to_check = ['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 
                    'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']

# Plotting boxplots for each numeric column to identify outliers
plt.figure(figsize=(8, 11))
for i, column in enumerate(columns_to_check, 1):
    plt.subplot(5, 3, i)
    sns.boxplot(x=df[column], color='orange')
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
plt.tight_layout()
plt.show()
```

### Remove outliers using IQR(Interquartile Range)

```python
# Remove outliers 
columns_to_check = ['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 
                    'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']
for i in columns_to_check:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1           # Calculate the Interquartile Range (IQR)

    # Define the lower and upper bounds for detecting outliers
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[i] <= upper_bound)]
```

```python
# after removing outlier
columns_to_check = ['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 
                    'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']

# Plotting boxplots for each numeric column to identify outliers
plt.figure(figsize=(5, 7))
for i, column in enumerate(columns_to_check, 1):
    plt.subplot(5, 3, i)
    sns.boxplot(x=df[column], color='orange')
    plt.title(f'Boxplot of {column}',fontsize =8)
    plt.xlabel(column,fontsize = 8)
plt.tight_layout()
plt.show()
```

```python
# Grouping the dataframe by 'Category' and calculating the mean calories for each category
top_5_meal_calories = df.groupby('Category')['Calories'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(6, 4))
top_5_meal_calories.plot(kind='bar',color='skyblue',label='category')
plt.title('Top 5 Categories for High Calories') 
plt.ylabel('Avg Calories')
plt.xlabel('Category') 

# Rotating the x-axis labels to 45 degrees for better readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Filtering the dataframe to find rows where 'Item' contains the word 'Grilled' 
grilled_chicken = df[df["Item"].str.contains('Grilled', case=False)]

# Filtering the dataframe to find rows where 'Item' contains the word 'Crispy' (case-insensitive)
crispy_chicken = df[df["Item"].str.contains('Crispy', case=False)]

# Creating a new dataframe to calculate the average values for Carbohydrates, Protein, and Total Fat
avg_nutrition = pd.DataFrame({
    'Grilled Chicken': grilled_chicken[['Carbohydrates', 'Protein', 'Total Fat']].mean(),
    'Crispy Chicken': crispy_chicken[['Carbohydrates', 'Protein', 'Total Fat']].mean()    
})

# Plotting a bar chart to visualize the nutritional differences between grilled and crispy chicken
avg_nutrition.plot(kind='bar',figsize=(6, 4),color=['salmon', 'lightgreen'])
plt.title('Nutritional Difference: Grilled & Crispy Chicken')
plt.xlabel('Nutrients')
plt.ylabel('Avg Amount')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
 # Histogram of Calories
plt.figure(figsize=(8, 4))
plt.hist(df['Calories'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Calories in Menu Items')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
# Define protein categories based on arbitrary ranges
def protein_category(protein):
    if protein < 10:
        return 'Low'
    elif 10 <= protein < 20:
        return 'Medium'
    else:
        return 'High'

# Apply the function to categorize items
df['Protein Category'] = df['Protein'].apply(protein_category)

# Plot a pie chart of the protein categories
protein_counts = df['Protein Category'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(protein_counts, labels=protein_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'gold', 'lightcoral'])
plt.title('Proportion of Protein Categories in Menu Items')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x='Calories', y='Total Fat', data=df, hue='Protein', palette='coolwarm')
plt.title('Calories vs. Total Fat')
plt.xlabel('Calories')
plt.ylabel('Total Fat (g)')
plt.show()
```

```python
# check correlation 
corr_matrix = df[['Calories', 'Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']].corr()

# Plot a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Nutritional Values')
plt.show()
```
